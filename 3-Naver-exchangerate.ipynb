```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
# This is to call the information of the website
# We are looking for exchange rate Won to dollar
url = 'https://finance.naver.com/marketindex/'
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
print(soup.prettify())
```

```python
# This to return the value has 'span' and 'value', which is located column [0]
soup.find_all('span', 'value')[0].string
```
